# Q1. What is Linux
Ans:
* It is an operating system, by using that, users/applications can communicate with hard ware components.
* It was developed/created in 1960s.
* With lot of extensions and improvements to base version, several flavours introduced by organization/companies (flavours like Redhat linux, ubuntu, CentOS etc)
**Linux is a *free and open-source operating system** that is based on **Unix**. It is widely used in servers, desktops, mobile devices (like Android), and embedded systems.

###  Key Features of Linux:

* **Open Source**: Anyone can view, modify, and distribute the source code.
* **Multi-user and Multi-tasking**: Multiple users can use the system, and many tasks can run at once.
* **Secure**: Linux is known for its strong user permissions and security model.
* **Portable**: Can run on many types of hardware, from phones to supercomputers.
* **Customizable**: You can modify everything from the kernel to the desktop environment.

---

###  Linux Structure:

1. **Kernel** – The core of the OS that communicates with hardware.
2. **Shell** – Interface between user and kernel (command-line).
3. **File System** – Organized structure for storing files and directories.
4. **Utilities** – Programs and commands for managing the system.

---

###  Common Linux Distributions (Distros):

* **Ubuntu** – User-friendly, great for beginners.
* **Debian** – Stable and versatile.
* **Fedora** – Cutting-edge features.
* **Arch Linux** – Lightweight and customizable.
* **CentOS/Red Hat** – Used in enterprise environments.

---

###  Where Is Linux Used?

* Servers (web, database, file servers)
* Android phones (Android uses a Linux kernel)
* IoT devices
* Supercomputers
* Personal desktops (by power users and developers)


# Q2. What is the difference between Hard Link & Soft Link?
Ans:
In **Linux**, **hard links** and **soft links (symbolic links)** are two ways of creating references to files.


###  **Hard Link**

A hard link is **another name for the same file**. Both the original file and the hard link point to the **same inode (data block on disk)**.

####  Characteristics:

* Points directly to the **same data/inode**.
* If the **original file is deleted**, the hard link still **retains the data**.
* Cannot link to **directories** (for safety).
* Only works **within the same file system**.

####  Example:

```bash
ln original.txt hardlink.txt
```

---

###  **Soft Link (Symbolic Link)**

A soft link is like a **shortcut**. It is a separate file that points to the **name/path** of the original file.

####  Characteristics:

* Points to the **filename/path**, not the data.
* If the **original file is deleted**, the soft link becomes **broken (dangling)**.
* Can link to **directories**.
* Can work **across file systems**.

####  Example:

```bash
ln -s original.txt softlink.txt
```

---

###  Summary Table:

| Feature              | Hard Link        | Soft Link (Symbolic)     |
| -------------------- | ---------------- | ------------------------ |
| Points to            | File's **inode** | File **name/path**       |
| File systems         | Same only        | Across file systems      |
| Survives deletion?   | ✅ Yes            | ❌ No (becomes broken)    |
| Links to directories | ❌ No             | ✅ Yes                    |
| Command              | `ln`             | `ln -s`                  |
| Size (of link file)  | Same as original | Small (just stores path) |


# Q3. What is a Kernel in Linux?
Ans:
The **Kernel** in Linux is the **core component** of the operating system. It acts as a **bridge between the hardware and software**.

---

###  What Does the Kernel Do?

1. **Manages Hardware**: Handles communication between software and hardware devices (CPU, memory, disk, etc.).
2. **Resource Allocation**: Distributes CPU time, memory, and device access among processes.
3. **Process Management**: Starts, schedules, and stops programs (called processes).
4. **Memory Management**: Controls how memory is used and shared.
5. **Device Drivers**: Interfaces with hardware components through drivers.
6. **System Calls**: Provides an interface for applications to request services (e.g., read a file, send data).

---

###  Role of Kernel in Linux System Structure:

```
+--------------------------+
|     User Applications    |
+--------------------------+
|         Shell            |   <-- Command interface
+--------------------------+
|         Kernel           |   <-- Manages system resources
+--------------------------+
|      Hardware (CPU, RAM, I/O) |
+--------------------------+
```

---

###  Types of Linux Kernels:

* **Monolithic Kernel** (used by Linux): All core functions (file systems, memory, devices) run in one large process.
* **Modular Design**: Linux kernel can load/unload parts (modules) as needed (e.g., device drivers).

---

###  Common Kernel Commands:

* View kernel version:

  ```bash
  uname -r
  ```
* Update kernel (Ubuntu example):

  ```bash
  sudo apt update && sudo apt upgrade
  ```

---

###  Summary:

| Feature     | Description                          |
| ----------- | ------------------------------------ |
| Type        | Monolithic + Modular                 |
| Main Role   | Bridge between software and hardware |
| Accessed by | System calls and drivers             |
| Manages     | CPU, memory, I/O, files, processes   |



# Q4. How do you create a user account?
Ans:
To **create a user account** in **Linux**, you use the `useradd` or `adduser` command (depending on the distribution).

---

###  1. Using `useradd` (Low-level command)

```bash
sudo useradd -m username
```

####  Explanation:

* `sudo` – Run as root (admin)
* `useradd` – Command to add user
* `-m` – Creates a home directory (e.g., `/home/username`)
* `username` – The new user's name

To set a password:

```bash
sudo passwd username
```

---

###  2. Using `adduser` (Interactive, user-friendly)

```bash
sudo adduser username
```

This command:

* Creates the user
* Sets up home directory
* Prompts for password and user details

---

###  Optional: Add User to a Group

```bash
sudo usermod -aG groupname username
```

Example (add to `sudo` group to give admin rights):

```bash
sudo usermod -aG sudo username
```

---

###  Verify the User

```bash
id username         # Shows user ID and group info
cat /etc/passwd     # Lists all users
```

# Q5. What is the ‘grep’ command used for in Linux?
Ans:
The **`grep`** command in Linux is used to **search for patterns or text** within files. It stands for **Global Regular Expression Print**.

---

###  Basic Syntax:

```bash
grep [options] pattern [file...]
```

---

###  Common Use Cases:

| Command                   | Description                                    |                                            |
| ------------------------- | ---------------------------------------------- | ------------------------------------------ |
| `grep "text" file.txt`    | Search for `"text"` in `file.txt`.             |                                            |
| `grep -i "text" file.txt` | Case-insensitive search.                       |                                            |
| `grep -r "text" /path`    | Recursively search in all files under `/path`. |                                            |
| `grep -n "text" file.txt` | Show line numbers with matches.                |                                            |
| `grep -v "text" file.txt` | Show lines **that do not match**.              |                                            |
| \`grep -E "a              | b"\`                                           | Use **extended regex** (like OR patterns). |

---

###  Example:

Given a file `notes.txt`:

```
Linux is fast.
LINUX is powerful.
I like Unix.
```

```bash
grep -i "linux" notes.txt
```

**Output:**

```
Linux is fast.
LINUX is powerful.
```

---

###  With Other Commands:

```bash
ps aux | grep firefox       # Find running processes with "firefox"
dmesg | grep -i usb         # Search USB messages in kernel log
```

---

###  Summary:

* **Searches text using patterns**
* **Supports regular expressions**
* **Works with files or command output**
